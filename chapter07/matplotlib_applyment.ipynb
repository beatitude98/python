```python
import pandas as pd
import matplotlib.pyplot as plt

daejeon = pd.read_csv('daejoen_temperature.csv', encoding='cp949')
daejeon['일시'] = pd.to_datetime(daejeon['일시'])
daejeon.set_index('일시', inplace=True)

june = daejeon[daejeon.index.month == 6]
july = daejeon[daejeon.index.month == 7]
august = daejeon[daejeon.index.month == 8]

avg6 = june.groupby(june.index.year)['평균기온(°C)'].mean()
avg7 = july.groupby(july.index.year)['평균기온(°C)'].mean()
avg8 = august.groupby(august.index.year)['평균기온(°C)'].mean()


plt.rc('font', family='NanumGothic')
plt.figure(figsize=(6.4, 4.8))
plt.title('2013~2022 대전시 6월, 7월, 8월 평균기온 변화')

plt.plot(avg6.index, avg6.values, label='6월 평균기온(°C)', color='r')
plt.plot(avg7.index, avg7.values, label='7월 평균기온(°C)', color='g')
plt.plot(avg8.index, avg8.values, label='8월 평균기온(°C)', color='b')

plt.xlabel('년도')
plt.ylabel('평균기온(°C)')
plt.legend()
plt.show()
```

```python
# 위 2개 묶는 연습
import pandas as pd
import matplotlib.pyplot as plt

daejeon = pd.read_csv('daejoen_temperature.csv', encoding='cp949')
daejeon['일시'] = pd.to_datetime(daejeon['일시'])
daejeon.set_index('일시', inplace=True)

plt.rc('font', family='NanumGothic')
plt.figure(figsize=(15, 8))
plt.plot(daejeon['평균기온(°C)'], label='평균기온', color='y')

plt.xlabel('년도')
plt.ylabel('평균온도')
plt.title('2013~2022 대전시의 기온 변화')

june = daejeon[daejeon.index.month == 6]
july = daejeon[daejeon.index.month == 7]
august = daejeon[daejeon.index.month == 8]

avg6 = june.groupby(june.index.year)['평균기온(°C)'].mean()
avg7 = july.groupby(july.index.year)['평균기온(°C)'].mean()
avg8 = august.groupby(august.index.year)['평균기온(°C)'].mean()

plt.plot(avg6.index, avg6.values, label='6월 평균기온(°C)', color='r')
plt.plot(avg7.index, avg7.values, label='7월 평균기온(°C)', color='g')
plt.plot(avg8.index, avg8.values, label='8월 평균기온(°C)', color='b')

plt.xlabel('년도')
plt.ylabel('평균기온(°C)')

plt.legend()
plt.show()
```

```python
df = pd.read_csv('daejoen_temperature.csv', encoding='cp949')
df['일시'] = pd.to_datetime(df['일시'])
df['일시']=pd.to_datetime(df['일시'],format='%Y-&m-&d')
df.info()
```

```python
group_by_year=df.groupby(df['일시'].dt.year)['평균기온(°C)'].mean() #mean 평균값 구하는것
#dt는 데이터프레임 열의 날짜 및 시간 관련 기능을 사용하기 위한 접근자
group_by_year

plt.rc('font',family="NanumGothic")
plt.rc('axes',unicode_minus="False")
plt.plot(group_by_year, 'b') #b 는 컬러명
plt.xlabel('연도')
plt.ylabel('평균기온(°C)')
# plt.legend()
plt.show()
```

```python
df.head()
```

```python
m = df['일시'].dt.month
df_summer = df[(m>=6)&(m<=8)]
group_by_summer = df_summer.groupby(df_summer['일시'].dt.year)['평균기온(°C)'].mean()
group_by_summer
```

```python
plt.plot(group_by_summer,'r')
plt.xlabel('연도')
plt.ylabel('평균온도')
plt.title('2013~2022년 대전 지역 여름철 (6~8월) 평균 온도 변화')
plt.show()
```

```python
df_june = df[df['일시'].dt.month == 6]
df_july = df[df['일시'].dt.month == 7]
df_august = df[df['일시'].dt.month == 8]
group_by_june = df_june.groupby(df_june['일시'].dt.year)['평균기온(°C)'].mean()
group_by_july = df_july.groupby(df_july['일시'].dt.year)['평균기온(°C)'].mean()
group_by_august = df_august.groupby(df_august['일시'].dt.year)['평균기온(°C)'].mean()

plt.plot(group_by_june, label='6월', color='r')
plt.plot(group_by_july, label='7월', color='g')
plt.plot(group_by_august, label='8월', color='b')
plt.legend()
```
