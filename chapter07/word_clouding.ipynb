```python
import matplotlib.pyplot as plt
#Natural Language Processing
from wordcloud import WordCloud
from konlpy.tag import Okt  # ==> import konlpy.tag.Okt
from collections import Counter  #==> import collections.Counter
import numpy as np

with open('대한민국헌법.txt', 'r', encoding="utf-8") as f:
    text = f.read()
    
okt = Okt()
nouns = okt.nouns(text)
# words = []
# for n in nouns:
#     if len(n)> 1:
#         words.append(n)


words = [n for n in nouns if len(n) > 1]
c = Counter(words)

#워드클라우드 생성하기
wc = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumGothic.ttf', width=2000, height=800, scale=2, max_font_size=250, max_words=20, background_color='white')
gen = wc.generate_from_frequencies(c)

plt.figure()
plt.imshow(gen)
wc.to_file('대한민국헌법.png')
```
