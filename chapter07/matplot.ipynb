```python
import matplotlib.pyplot as plt

# plt.plot([x값 리스트,]y값리스트[,포맷 스타일])
plt.title("대한민국")
plt.plot([-1,2,3,4]) #이건 y값리스트 x는 생략이 가능
plt.rcParams['font.family'] = 'NanumGothic'
plt.show()
```

```python
year =[2014, 2017, 2020, 2023]
price = [25000, 31000, 53000, 63000]
plt.plot(year, price, 'rs:')

plt.show()
```

```python
year =[2014, 2017, 2020, 2023]
price = [25000, 31000, 53000, 63000]
price2 = [22000, 36000, 58000, 65000]
plt.plot(year, price, 'rs:')
plt.plot(year, price2, 'b+-')
plt.axis([2013, 2024, 20000, 70000])
plt.show()
```

```python
# 차트 하나에 그래프 3개생성
import numpy as np
plt.plot(np.random.randn(10), 'k', label='one')
plt.plot(np.random.randn(10) * 4, 'r--', label='two')
plt.plot(np.random.randn(10) * 10, 'g-', label='three')
plt.legend() #legend: 범례 => 레이블(label) 옵션으로 지정한 값이 표현됨
plt.show()
```

```python
# 다중 그래프 (그래프를 한꺼번에 여러개 그리기)
plt.subplot(2,2,1)
plt.plot(np.random.randn(10), 'b--')
plt.subplot(2,2,2)
plt.plot(np.random.randn(100),'r',alpha=0.7)
plt.subplot(2,2,3)
plt.plot(np.random.randn(10),'y^')
plt.subplot(2,2,4)
plt.plot(np.random.randn(10),'g.')
plt.show()
```

```python
# 산점도 Scatter
year =[2014, 2017, 2020, 2023]
price = [25000, 31000, 53000, 63000]
plt.scatter(year, price, c='g', s=40, marker='^')
plt.axis([2013, 2024, 20000, 70000])
plt.show()
```

```python
# 히스토그램(histogram): 연속할 변수의 구간별 분포
# 주사위를 던졌을때 1부터 6까지 수가 몇번씩 나왔는지
numbers = [5, 4, 4, 1, 6, 3, 4, 1, 2, 2]
plt.hist(numbers, bins=6)
plt.show()
```
