```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('daejeon_1904_2023.csv', encoding="cp949")
df
```

```python
dw=df.groupby(df['일시'])['평균기온(°C)'].mean() #mean 평균값 구하는것
dw

plt.rc('font',family="NanumGothic")
plt.rc('axes',unicode_minus="False")
plt.plot(dw, 'b')
plt.xlabel('연도')
plt.ylabel('평균기온(°C)')
```
