```python
# 막대 그래프 bar plot
# 국어 영어 수학 점숨가 각각 85 76 55일때 막대 그래프를 작성해보자
import matplotlib.pyplot as plt

subject = ['국어','영어', '수학']
grade = [85,76,55]
plt.title('점수보고서')
plt.xlabel('과목')
plt.ylabel('점수')
plt.bar(subject, grade)
plt.ylim(0,100)
plt.text(0,90,'Good')
plt.text(2,60,'Bad!')
plt.show()
```

```python
g1 = nx.Graph()
g1.add_node('David')
g1.add_node('John')
g1.add_node('Mary')
g1.add_node('Julia')
g1.add_node('Juda')
g1.add_nodes_from(['Judy','Karen'])

g1.add_edge('David','John')
g1.add_edge('David','Mary')
g1.add_edge('David','Karen')
g1.add_edges_from([('John','Mary'),('John','Judy')])
g1.remove_edge('Mary','John')

print(g1.degree)

nx.draw(g1, nodelist=dict(g1.degree).keys(), node_size=[v * 1000 for v in dict(g1.degree).values()], with_labels=True, font_weight='bold')
```

```python
#가중치(Weight)
import networkx as nx
g2 = nx.DiGraph()
g2.add_weighted_edges_from([(1,2,10), (2,3,20)])
g2.add_edge(1,3,weight=30)
# 그래프를 시각화할 때 노드 간 거리를 최대한 가깝게 유지하면서 시각적으로 균형잡힌 레이아웃을 만들 때 사용하는 옵션
pos = nx.spring_layout(g2)

nx.draw(g2, pos=pos, with_labels=True)
label = nx.get_edge_attributes(g2, 'weight')
nx.draw_networkx_edge_labels(g2, pos, edge_labels = label)
```
