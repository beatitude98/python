```python
#문제 : 올해를 기준으로 대전의 지난 40년 간의 평균기온, 최저기온, 최고기온의 변화를 시각화하시오(꺽은선 그래프)

import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('daejeon_temp.csv', encoding='cp949')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df2 = df.bfill()
df2
```

```python
plt.rc('font', family='NanumGothic')
plt.figure(figsize=(15,8))
plt.xticks(range(1, len(df2) + 1), df2['일시'])
plt.plot(range(1, len(df2) +1), df2['평균기온(°C)'], label='평균기온',c='g')
plt.plot(range(1, len(df2) +1), df2['최고기온(°C)'], label='최고기온',c='r')
plt.plot(range(1, len(df2) +1), df2['최저기온(°C)'], label='최저기온',c='b')
plt.xlabel('년도')
plt.ylabel('기온')
plt.legend()
plt.show()
```
