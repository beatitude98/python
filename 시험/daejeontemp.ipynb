```python
#문제 : 올해를 기준으로 대전의 지난 40년 간의 평균기온, 최저기온, 최고기온의 변화를 시각화하시오(꺽은선 그래프)

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('daejeon_temp.csv', encoding="cp949")
df
```

```python
dw=df.groupby(df['일시'])['평균기온(°C)'].mean() #mean 평균값 구하는것
dw

plt.rc('font',family="NanumGothic")
plt.rc('axes',unicode_minus="False")
plt.plot(dw, 'b')
plt.xlabel('연도')
plt.ylabel('평균기온(°C)')
```

```python
plt.rc('font',family="NanumGothic")
plt.rc('axes',unicode_minus="False")
plt.ylabel('평균온도')
df2 = df.set_index(df['일시'])
df2 = df['평균기온(°C)']
df2.plot(c= 'b')
```
