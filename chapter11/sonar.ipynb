```python
import pandas as pd
from sklearn.model_selection import  train_test_split
from keras.models import Sequential
from keras.layers import Dense
import pandas as pd

df_sonar = pd.read_csv('sonar3.csv', encoding='utf-8', header=None)
df_sonar
```

```python
X = df_sonar[df_sonar.columns[:60]]
y = df_sonar[df_sonar.columns[60:]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = Sequential()
model.add(Dense(60, input_dim=60, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(7, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics='accuracy')

model.fit(X_train, y_train, epochs=200, batch_size=10)

model.evaluate(X_test, y_test)
```

```python
from keras.callbacks import EarlyStopping

early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20)

model.fit(X_train, y_train, validation_split=0.25 ,epochs=100, batch_size=10, callbacks=early_stopping_callback)

model.evaluate(X_test, y_test)
```
