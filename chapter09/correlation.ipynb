```python
import pandas as pd

eng_listening = [30, 60, 90]
eng_score = [70, 80, 90]

data = {'eng_listening' : eng_listening, 'eng_score' : eng_score}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df['eng_listening'],df['eng_score'],'r')
```

```python
coef = df.corr(method='pearson')
coef
```

```python
eng_listening = [30, 60, 90, 31, 32, 69, 92 ,99]
eng_reading = [40, 45, 60, 20, 15, 70, 60, 80]
eng_class = [60, 120, 120, 60, 60, 180, 120, 120]
eng_score = [70, 80, 90, 70, 71, 85, 90, 92]

data3 = {'eng_listening' : eng_listening ,'eng_reading': eng_reading,'eng_class' : eng_class, 'eng_score' : eng_score }
df3 = pd.DataFrame(data3)
```

```python
pearson_coef = df3.corr(method='pearson')
print(pearson_coef)

spearman_coef = df3.corr(method='spearman')
print(spearman_coef)

kendall_coef = df3.corr(method='kendall')
print(kendall_coef)
```
