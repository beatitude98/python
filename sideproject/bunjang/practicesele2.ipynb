```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

url = "https://m.bunjang.co.kr/"

driver = webdriver.Chrome()
driver.get(url)

# 입력 상자를 찾기 위해 WebDriverWait를 사용합니다.
search_box = WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="root"]/div/div/div[2]/div[1]/div[1]/div[1]/div[1]/input'))
)

# 입력 상자에 '크리드'를 입력하고 Enter 키를 누릅니다.
search_box.send_keys('크리드')
search_box.send_keys(Keys.RETURN)

# 상품 리스트를 기다립니다. 여기서는 10초 동안 기다립니다.
product_list = WebDriverWait(driver, 10).until(
    EC.presence_of_all_elements_located((By.CLASS_NAME, "sc-cTjmhe"))
)

for i in range(1, 101):
    xpath = f'//*[@id="root"]/div/div/div[4]/div/div[4]/div/div[{i}]/a/div[2]/div[1]'
    
    try:
        # 상품 이름
        product_name = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, xpath))
        ).text
        print(f'상품명 {i}: {product_name}')
    except Exception as e:
        print(f'상품명 {i}를 가져오는 데 실패했습니다. 오류 메시지: {e}')
        
for j in range(1, 101):
    xpath = f'//*[@id="root"]/div/div/div[4]/div/div[4]/div/div[{j}]/a/div[2]/div[2]/div[1]'

    try:
        # 상품 가
        product_price = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, xpath))
        ).text
    #     print(f'가격 : {j} : {product_price}')
    # except Exception as e:
    #     print(f'가격 {j}을 가져오는 데 실패했습니다. 오류 메시지: {e}')
        
        print(f'상품명: {product_name} : 가격 {product_price}')
    except Exception as e:
        print(f'상품명 {i}를 가져오는 데 실패했습니다. 오류 메시지: {e}')
```
